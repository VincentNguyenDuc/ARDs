```python
# import libraries
import regression as rd
import sys
import matplotlib.pyplot as plt
import numpy
%matplotlib inline
```

```python

#* Set up

def truefunc(x):
    return numpy.sin(x*2.0)*numpy.sqrt(x)/3.3

ydisplaymin = -1.8
ydisplaymax = 1.8
points = 30
rd.setSigma(0.2)


#*  generate random points

# for points
x = rd.makeX(points)
yt = truefunc(x)

# for curve
xts = rd.makeX(200, uniform=True)
yts = truefunc(xts)

# *  adds Gaussian noise
y = rd.addNoise(yt)
```

```python

#* build the poly curve from order-th regression
order = 11
ys = rd.linReg(x, y, xts, order)


#* plot data and the truth
plt.plot(xts, yts, label='truth', linewidth=2)
plt.plot(x, y, 'o')
plt.ylabel('y')
plt.xlabel('x')


#* plot fitted curve
plt.plot(xts, ys, label='poly ' + str(order), linewidth=1.25)
plt.ylim(ydisplaymin, ydisplaymax)
plt.legend()
plt.title('A 11-th polynomial fit')
```
